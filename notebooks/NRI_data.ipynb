```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nodes = np.load("data/loc_train_springs5_short.npy")
velocities = np.load("data/vel_train_springs5_short.npy")
edges = np.load("data/edges_train_springs5_short.npy")
```

```python
print(nodes.shape) # simulation_index, timesteps within the simulation (5000 timesteps sampled every 100 steps - 1 = 49 total) , x & y positions, node_index
print(velocities.shape) # simulation_index, timesteps within the simulation (5000 timesteps sampled every 100 steps - 1 = 49 total), x & y velocities, node_index
print(edges.shape) # simulation_index, node_index_start, node_index_end. This is fixed for the whole 49-step sequence. Note that the entries are either 0 or 1 -- 2 types of springs (i.e. spring or no spring)
```

```python
fig, ax = plt.subplots(1,3, figsize=(21,7))
fig.suptitle("Spring Simulations", fontsize=30)

for i in range(3):
    sim_number = i
    ax[i].plot(nodes[sim_number, :, 0, 0], nodes[sim_number, :, 1, 0], marker='o', markevery=6, label="Ball 0")
    ax[i].plot(nodes[sim_number, :, 0, 1], nodes[sim_number, :, 1, 1], marker='o', markevery=6, label="Ball 1")
    ax[i].plot(nodes[sim_number, :, 0, 2], nodes[sim_number, :, 1, 2], marker='o', markevery=6, label="Ball 2")
    ax[i].plot(nodes[sim_number, :, 0, 3], nodes[sim_number, :, 1, 3], marker='o', markevery=6, label="Ball 3")
    ax[i].plot(nodes[sim_number, :, 0, 4], nodes[sim_number, :, 1, 4], marker='o', markevery=6, label="Ball 4")
    ax[i].set_title("Simulation #{}".format(sim_number), fontsize=14)
    ax[i].tick_params(labelsize=12)
    ax[i].legend(fontsize=16)
```
